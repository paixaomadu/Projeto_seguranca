```python
import pandas as pd
import matplotlib.pyplot as plt
```

#Tratamento de dados

```python
df_filmes = pd.read_csv('world_imdb_movies_top_movies_per_year.csv', sep = ',')
df_filmes.head()
```

#Removendo colunas

```python
df_filmes.drop(['budget', 'gross_world_wide', 'link', 'id', 'rating_mpa','writer','star', 'filming_location','production_company','win', 'nomination', 'oscar', 'director'], axis = 1, inplace = True)
df_filmes.head()
```

#Dividindo a coluna "Genre" em Dez colunas correspondendo aos Gêneros separadamente

```python
df_filmes[['Genero_1', 'Genero_2', 'Genero_3', 'Genero_4', 'Genero_5', 'Genero_6', 'Genero_7', 'Genero_8', 'Genero_9', 'Genero_10']] = df_filmes['genre'].str.split(',', expand = True)
df_filmes.head()
```

#Trazendo o resultado de Gêneros com maior nota

```python
df_filmes_media_genero = df_filmes.groupby(['Genero_1'])['rating_imdb'].mean().reset_index()
df_filmes_media_genero = df_filmes_media_genero.sort_values(by='rating_imdb', ascending = False)
df_filmes_media_genero = df_filmes_media_genero.head()
```

```python
df_filmes_media_genero.plot(kind='barh', x='Genero_1', y='rating_imdb')
plt.xlabel('Média de notas')
plt.ylabel('Gêneros')
plt.title('Gênero com maior nota')
plt.show()
df_filmes_media_genero.head()
```

#trazendo o resultado de filme com maior nota

```python
df_filmes_maior_nota = df_filmes.groupby(['title'])['rating_imdb'].max().reset_index()
df_filmes_maior_nota = df_filmes_maior_nota.sort_values(by='rating_imdb', ascending = False)
df_filmes_maior_nota = df_filmes_maior_nota.head()
```

```python
df_filmes_maior_nota.plot(kind='barh', x='title', y='rating_imdb')
plt.xlabel('Título do filme')
plt.ylabel('Notas')
plt.title('O filme com maior nota')
plt.show()
df_filmes_maior_nota.head()
```

#Trazendo o resultado de filmes com mais votos

```python
df_filmes_mais_votos = df_filmes.groupby(['title'])['vote'].max().reset_index()
df_filmes_mais_votos = df_filmes_mais_votos.sort_values(by='vote', ascending = False)
df_filmes_mais_votos = df_filmes_mais_votos.head()
```

```python
df_filmes_mais_votos.plot(kind='barh', x='title', y='vote')
plt.xlabel('Votos')
plt.ylabel('Título do filme')
plt.title('O filme com mais votos')
plt.show()
df_filmes_mais_votos.head()
```

```python
df_filmes_pais = df_filmes.groupby(['country_origin']).size().reset_index(name='Quantidade')
df_filmes_pais = df_filmes_pais.sort_values(by='Quantidade', ascending = False)
df_filmes_pais = df_filmes_pais.head()
```

```python
df_filmes_genero = df_filmes.groupby(['Genero_1']).size().reset_index(name='Quantidade')
df_filmes_genero = df_filmes_genero.sort_values(by='Quantidade', ascending = False)
df_filmes_genero=df_filmes_genero.head()
```

#Tazendo o resultado de gêneros mais presentes no DataSet

```python
df_filmes_genero.plot(kind='pie', x='title', y='Quantidade')
plt.xlabel('Gênero')
plt.ylabel('Quantidade')
plt.title('Gêneros mais presentes no DataSet')
plt.show()
df_filmes_genero.head()
```
